#### Normalization- min max scaler

```python
import seaborn as sns
```

```python
df=sns.load_dataset('taxis')
```

```python
df.head()
```

```python
distance=list(df['distance'])
```

```python
distance
```

```python
sns.histplot(distance)
```

```python
import numpy as np
min=np.min(distance)
max=np.max(distance)
```

```python
min,max
```

```python
normalized_distance=[]
for i in distance: 
    Zscaled=(i-min)/(max-min)
    normalized_distance.append(Zscaled)

print(normalized_distance)
```

```python
sns.histplot(normalized_distance)
```
