#### Nominal Encoding or One Hot Encoding (OHE)

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'ID':[5,6,7,8,9,10],'color': ['red','blue','green','green','red','blue']})
```

```python
df
```

```python
# create a instance of one hot encoder
encoder=OneHotEncoder()
```

```python
# fit the encoder to the dataframe and transform the categorical variable
encoded=encoder.fit_transform(df[['color']])           
```

```python
encoded   # sparce matrix
```

```python
encoded.toarray()       # convert matrix into array
```

```python
encoder.get_feature_names_out()
```

```python
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
df            # this is origional data
```

```python
OHE_data=pd.concat([df[['ID']],encoded_df],axis=1)    # this is the encoded data set
```

```python
OHE_data
```
